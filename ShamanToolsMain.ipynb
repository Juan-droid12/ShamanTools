```python
# Base de datos de usuarios y productos
usuarios = [
    {"ID": 1, "nombre_usuario": "Ren Tao", "contraseña": "1234_4321", "rol": "admin"},
    {"ID": 2, "nombre_usuario": "Ioh Asakura", "contraseña": "abc_cba", "rol": "vendedor"},
    {"ID": 3, "nombre_usuario": "Chocolove McDonnell", "contraseña": "11110000", "rol": "vendedor"}
]

productos = [
    {"ID": 1, "nombre_producto": "Ladrillo 10x20x30", "precio_compra": 1200.0, "precioVenta": 1500.0, "cantidadProducto": 1000},
    {"ID": 2, "nombre_producto": "Cemento Gris x50kg", "precio_compra": 28000.0, "precioVenta": 30000.0, "cantidadProducto": 200},
    {"ID": 3, "nombre_producto": "Varilla Corrugada #5 x6m", "precio_compra": 33000.0, "precioVenta": 35000.0, "cantidadProducto": 150},
    {"ID": 4, "nombre_producto": "Impermeabilizante 22kg", "precio_compra": 350000.0, "precioVenta": 360000.0, "cantidadProducto": 50},
    {"ID": 5, "nombre_producto": "Malla Electro Soldada 6x2.2m", "precio_compra": 42000.0, "precioVenta": 45000.0, "cantidadProducto": 75}
]

ventas = []

# Función para mostrar el mensaje de bienvenida
def mostrar_bienvenida():
    print("Bienvenido/a al Sistema de Inventario de Shaman Tools")

# Función para iniciar sesión
def iniciar_sesion():
    nombre_usuario = input("Ingrese su nombre de usuario: ")
    contraseña = input("Ingrese su contraseña: ")

    for usuario in usuarios:
        if nombre_usuario == usuario["nombre_usuario"] and contraseña == usuario["contraseña"]:
            print(f"Bienvenido/a, {nombre_usuario}")
            return usuario["rol"], usuario["ID"]

    print("Usuario o contraseña incorrectos")
    return None, None

# Función para mostrar el menú del administrador
def menu_admin():
    print("\nOpciones de administrador:")
    print("1. Registrar nuevo usuario")
    print("2. Agregar o actualizar producto")
    print("3. Consultar inventario")
    print("4. Comprar producto para tienda")
    print("5. Ver resumen de ventas")
    print("6. Cerrar sesión")
    opcion_admin = input("Seleccione una opción: ")

    if opcion_admin == "1":
        registrar_usuario()
    elif opcion_admin == "2":
        agregar_producto()
    elif opcion_admin == "3":
        consulta_inventario()
    elif opcion_admin == "4":
        comprar_producto()
    elif opcion_admin == "5":
        ver_resumen_ventas()
    elif opcion_admin == "6":
        print("Cerrando sesión de administrador...")
    else:
        print("Opción no válida")

# Función para registrar un nuevo usuario
def registrar_usuario():
    nuevousuario = input("Ingrese el nombre del nuevo usuario: ")
    nuevacontraseña = input("Ingrese la contraseña del nuevo usuario: ")
    nuevorol = input("Ingrese el rol del nuevo usuario (admin/vendedor): ")
    nuevoID = len(usuarios) + 1
    usuarios.append({"ID": nuevoID, "nombre_usuario": nuevousuario, "contraseña": nuevacontraseña, "rol": nuevorol})
    print(f"Usuario registrado exitosamente con ID: {nuevoID}")

# Función para agregar o actualizar productos
def agregar_producto():
    nombre_producto = input("Ingrese el nombre del producto: ")
    precio_compra = float(input("Ingrese el precio de compra: "))
    precioVenta = float(input("Ingrese el precio de venta: "))
    cantidadProducto = int(input("Ingrese la cantidad en inventario: "))
    productoID = int(input("Ingrese el ID del producto: "))

    for producto in productos:
        if producto["ID"] == productoID:
            producto["nombre_producto"] = nombre_producto
            producto["precio_compra"] = precio_compra
            producto["precioVenta"] = precioVenta
            producto["cantidadProducto"] += cantidadProducto
            print("Producto actualizado exitosamente")
            return

    productos.append({"ID": productoID, "nombre_producto": nombre_producto, "precio_compra": precio_compra, "precioVenta": precioVenta, "cantidadProducto": cantidadProducto})
    print(f"Producto agregado exitosamente con ID: {productoID}")

# Función para mostrar el menú del vendedor
def menu_vendedor():
    print("\nOpciones de vendedor:")
    print("1. Consultar inventario")
    print("2. Realizar venta")
    print("3. Cerrar sesión")
    opcion_vendedor = input("Seleccione una opción: ")

    if opcion_vendedor == "1":
        consulta_inventario()
    elif opcion_vendedor == "2":
        vender_producto()
    elif opcion_vendedor == "3":
        print("Cerrando sesión de vendedor...")
    else:
        print("Opción no válida")

# Función para consultar inventario
def consulta_inventario():
    consulta = input("Ingrese el ID del producto para buscar o 'todos' para ver inventario completo: ")

    if consulta.lower() == "todos":
        print("\nInventario completo:")
        for producto in productos:
            print(f"ID: {producto['ID']} | Producto: {producto['nombre_producto']} | Precio Venta: ${producto['precioVenta']} | Cantidad: {producto['cantidadProducto']} {producto['nombre_producto'].split()[-1]}")
    else:
        try:
            productoID = int(consulta)
            encontrado = False
            for producto in productos:
                if producto["ID"] == productoID:
                    print(f"ID: {producto['ID']} | Producto: {producto['nombre_producto']} | Precio Venta: ${producto['precioVenta']} | Cantidad: {producto['cantidadProducto']} {producto['nombre_producto'].split()[-1]}")
                    encontrado = True
                    break
            if not encontrado:
                print("Producto no encontrado")
        except ValueError:
            print("ID de producto inválido")

# Función para realizar ventas
def vender_producto():
    try:
        productoID = int(input("Ingrese el ID del producto: "))
        cant_vendida = int(input("Ingrese la cantidad vendida: "))

        for producto in productos:
            if producto["ID"] == productoID:
                if producto["cantidadProducto"] >= cant_vendida:
                    producto["cantidadProducto"] -= cant_vendida
                    ventas.append({"productoID": productoID, "cantidadVendida": cant_vendida})
                    print("Venta registrada exitosamente")
                else:
                    print("insuficiente")
                return
        print("Producto no encontrado")
    except ValueError:
        print("Entrada inválida")

# Función para comprar productos (solo admin)
def comprar_producto():
    try:
        productoID = int(input("Ingrese el ID del producto para comprar: "))
        cantidad_a_comprar = int(input("Ingrese la cantidad a comprar: "))

        for producto in productos:
            if producto["ID"] == productoID:
                producto["cantidadProducto"] += cantidad_a_comprar
                print("Compra registrada exitosamente")
                return
        print("Producto no encontrado")
    except ValueError:
        print("Entrada inválida")

# Función para ver resumen de ventas (solo admin)
def ver_resumen_ventas():
    print("\nResumen de ventas realizadas:")
    if not ventas:
        print("No se han realizado ventas aún.")
    else:
        for venta in ventas:
            print(f"Producto ID: {venta['productoID']} | Cantidad vendida: {venta['cantidadVendida']}")

# Programa principal del sistema de inventario
def sistema_inventario():
    mostrar_bienvenida()

    while True:
        print("\n1. Ingresar al sistema")
        print("2. Salir")
        opcion_principal = input("Seleccione una opción: ")

        if opcion_principal == "1":
            rol, userID = iniciar_sesion()
            if rol == "admin":
                while True:
                    menu_admin()
                    if opcion_admin == "6":
                        break
            elif rol == "vendedor":
                while True:
                    menu_vendedor()
                    if opcion_vendedor == "3":
                        break
        elif opcion_principal == "2":
            print("Gracias por usar el sistema. ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

sistema_inventario()
```
